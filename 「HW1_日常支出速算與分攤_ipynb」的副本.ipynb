```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

```python
import pandas as pd
# read data and put it in a dataframe
# 在 google 工作表載入 gsheets
gsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/1lLcOA_CUkl6N0H6OiAUnQyBIUyRR45Q3CHSrPoK7vCE/edit?usp=sharing')
```

```python
# 從 gsheets 的 All-whiteboard-device 載入 sheets
sheets = gsheets.worksheet('記帳本').get_all_values()
# 將 sheets1 資料載入 pd 的 DataFrame 進行分析
df = pd.DataFrame(sheets[1:], columns=sheets[0])
# 取得最前面的5筆資料
df.head()
```

```python
import datetime
```

```python

# 讓使用者輸入資料
date_str = input("請輸入日期 (格式：YYYY-MM-DD): ")
# 檢查日期格式是否正確
datetime.datetime.strptime(date_str, '%Y-%m-%d')

time_str = input("請輸入時間 (格式：HH:MM): ")
# 檢查時間格式是否正確
datetime.datetime.strptime(time_str, '%H:%M')

sort = input("請輸入分類: ")
item = input("請輸入品項: ")
amount = float(input("請輸入金額: "))
person = input("請輸入付款人: ")
place = input("請輸入地點: ")
methods = input("請輸入支付方式: ")
remark = input("請輸入備註: ")
```

```python
import pandas as pd

# 創建一個新的 DataFrame 來代表你要新增的資料
new_data = pd.DataFrame([{
    '日期': date_str,
    '時間': time_str,
    '分類': sort,
    '品項': item,
    '金額': amount,
    '付款人': person,
    '地點': place,
    '支付方式': methods,
    '備註': remark
}])

# 使用 concat() 將新資料合併到舊的 df 中
df = pd.concat([df, new_data], ignore_index=True)

```

```python
df
```

```python
# 將 DataFrame 轉換成列表的列表 (list of lists)，這是 gspread 支援的資料格式
data_to_write = df.values.tolist()

# 第一步：取得工作表物件
worksheet = gsheets.worksheet('記帳本')
```

```python
data_to_write[-1]
```

```python
# 第二步：將資料寫入工作表
row_to_write = data_to_write[-1]

# 將 NaN 轉換為 None，因為 gspread/JSON 不支援 NaN
cleaned_row = [None if pd.isna(x) else x for x in row_to_write]

worksheet.append_row(values=cleaned_row, value_input_option='USER_ENTERED')
print("資料已成功寫入 Google 工作表！")
```
